```python
# 1a. Load the movie dataset into a Pandas DataFrame using the pd.read_csv() function.
import pandas as pd

df = pd.read_csv('tmdb_movies_data.csv')

# 1b. Explore the DataFrame using df.head(), df.info(), and df.describe().
# Display the first few rows to understand the structure of the data
df.head()
```

```python
# Get a summary of the dataframe, including data types and non-null values
df.info()
```

```python
# Get statistical summary for numerical columns
df.describe()
```

```python
# 1ci. Handle any missing values (e.g., using df.fillna()).

# Identify numeric and non-numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
non_numeric_cols = df.select_dtypes(exclude=['number']).columns

# Fill missing values for only numeric columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill missing values for non-numeric columns with the most frequent value (mode)
df[non_numeric_cols] = df[non_numeric_cols].fillna(df[non_numeric_cols].mode().iloc[0])

# Check for missing values after filling
print(df.isnull().sum())
```

```python
# 1cii. Convert data types as needed (e.g., using pd.to_numeric()).

# Convert 'release_year' and 'runtime' to numeric
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')
```

```python
# 2ai. Calculate the average runtime of movies in each genre using np.mean() and groupby operations.

import numpy as np

# Split genres into separate genres
df['genres'] = df['genres'].str.split('|')

# Calculate the average runtime per genre
average_runtime_by_genre = df.explode('genres').groupby('genres')['runtime'].mean()
average_runtime_by_genre
```

```python
# 2aii. Find the movie with the longest runtime using np.argmax().

longest_runtime_index = np.argmax(df['runtime'])
longest_runtime_movie = df.iloc[longest_runtime_index]
longest_runtime_movie
```

```python
# 2bi. Create a new column for "movie age" by calculating the difference between the current year and the release year.

current_year = 2025
df['movie_age'] = current_year - df['release_year']
```

```python
# 2bii. Filter the DataFrame to display movies released in a specific year range.

# Filter movies released between 2010 and 2015
filtered_movies = df[(df['release_year'] >= 2010) & (df['release_year'] <= 2015)]
filtered_movies.head()
```

```python
# 2biii. Analyze the relationship between runtime and rating using correlation (df.corr()).

# Calculate correlation between runtime and rating
correlation = df[['runtime', 'vote_average']].corr()
correlation
```

```python
# 2ci. Create a histogram to visualise the distribution of movie ratings using plt.hist().

import matplotlib.pyplot as plt

# Plot histogram for movie ratings
plt.hist(df['vote_average'], bins=20, color='cornflowerblue', edgecolor='black')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2cii. Generate a scatter plot to show the relationship between runtime and rating using plt.scatter().

plt.scatter(df['runtime'], df['vote_average'], color='royalblue', alpha=0.6)
plt.title('Relationship between Runtime and Rating')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Rating')
plt.show()
```

```python
# 2ciii. Create a bar chart to compare the average rating of movies in different genres using plt.bar().

# Split genres and calculate the average rating per genre
average_rating_by_genre = df.explode('genres').groupby('genres')['vote_average'].mean()

# Plot a bar chart
average_rating_by_genre.plot(kind='bar', figsize=(10, 6), color='steelblue', edgecolor='black')
plt.title('Average Movie Rating by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
# 3a. Use Matplotlib to create visualisations that highlight interesting patterns or insights from the data.
# 3b. Add titles, labels, and legends to make your visualisations informative.

# Calculate average rating per year
avg_rating_by_year = df.groupby('release_year')['vote_average'].mean()

# Plot the average ratings by release year
plt.figure(figsize=(10, 6))
plt.plot(avg_rating_by_year.index, avg_rating_by_year.values, color='darkblue', marker='o', linestyle='-', linewidth=2, markersize=5)
plt.title('Average Movie Rating by Year', fontsize=14)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sort movies by popularity and select the top 10
top_10_popular_movies = df[['original_title', 'popularity']].sort_values(by='popularity', ascending=False).head(10)

# Plot the top 10 popular movies
plt.figure(figsize=(10, 6))
plt.barh(top_10_popular_movies['original_title'], top_10_popular_movies['popularity'], color='midnightblue', edgecolor='black')
plt.title('Top 10 Most Popular Movies', fontsize=16)
plt.xlabel('Popularity', fontsize=12)
plt.ylabel('Movie Title', fontsize=12)
plt.gca().invert_yaxis()  # To display the highest popularity at the top
plt.tight_layout()
plt.show()
```

```python
# Plot a scatter plot between budget and revenue
plt.figure(figsize=(10, 6))
plt.scatter(df['budget'], df['revenue'], alpha=0.5, color='powderblue')
plt.title('Budget vs Revenue', fontsize=14)
plt.xlabel('Budget ($)', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Histogram for movie runtime distribution
plt.figure(figsize=(10, 6))
plt.hist(df['runtime'], bins=30, color='lightblue', edgecolor='black')
plt.title('Distribution of Movie Runtimes', fontsize=14)
plt.xlabel('Runtime (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Vote Count vs Vote Average
plt.figure(figsize=(10, 6))
plt.scatter(df['vote_count'], df['vote_average'], alpha=0.6, color='skyblue')
plt.title('Vote Count vs Vote Average', fontsize=14)
plt.xlabel('Vote Count', fontsize=12)
plt.ylabel('Vote Average', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Count movies released per year
movies_per_year = df['release_year'].value_counts().sort_index()

# Bar chart for movies released per year
plt.figure(figsize=(10, 6))
plt.bar(movies_per_year.index, movies_per_year.values, color='cadetblue')
plt.title('Number of Movies Released per Year', fontsize=14)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Group by movie age and calculate average vote rating
movie_age_rating = df.groupby('movie_age')['vote_average'].mean().reset_index()

# Line plot for Vote Average vs Movie Age
plt.figure(figsize=(10, 6))
plt.plot(movie_age_rating['movie_age'], movie_age_rating['vote_average'], marker='o', color='cornflowerblue')
plt.title('Vote Average vs Movie Age', fontsize=14)
plt.xlabel('Movie Age (Years)', fontsize=12)
plt.ylabel('Average Vote Rating', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Sort by revenue and get top 10 movies
top_revenue_movies = df.sort_values(by='revenue', ascending=False).head(10)

# Bar chart for top 10 revenue movies
plt.figure(figsize=(10, 6))
plt.bar(top_revenue_movies['original_title'], top_revenue_movies['revenue'] / 1e6, color='dodgerblue', edgecolor='black')
plt.title('Top 10 Movies by Revenue', fontsize=14)
plt.xlabel('Movie Title', fontsize=12)
plt.ylabel('Revenue (in millions)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
```
